```python
from langdetect import detect

import pandas as pd

# Load the CSV file
df = pd.read_csv('Pumadataset_aspects_category.csv')

# Function to detect language and convert to English if not 'en' or 'id'
def detect_and_convert_language(text):
    try:
        lang = detect(text)
        if lang not in ['en', 'id']:
            return 'en'  # Convert to English if language is not English or Indonesian
        else:
            return lang
    except:
        return 'en'  # Convert to English if language detection fails

# Apply the language detection and conversion function to the 'normalized_text' column
df['Language'] = df['lemmatized_text'].apply(detect_and_convert_language)

# Display the first few rows with detected and converted languages
print(df[['lemmatized_text', 'Language']].head())

# Count the occurrences of each distinct language after conversion
language_counts = df['Language'].value_counts()

# Count how many distinct languages are there after conversion
num_distinct_languages = len(language_counts)

# Display the counts of each distinct language after conversion and the total number of distinct languages
print("Distinct languages detected and their counts after conversion:")
print(language_counts)
print("\nTotal number of distinct languages detected after conversion:", num_distinct_languages)
```
