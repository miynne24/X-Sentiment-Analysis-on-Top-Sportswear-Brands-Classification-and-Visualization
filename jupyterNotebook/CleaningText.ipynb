```python
import pandas as pd
```

```python
df=pd.read_csv("kkmart.csv")
```

```python
df
```

```python
#rename Multiple Columns in a list
df = df.rename(columns={'created_at':'date_created', 'full_text':'tweets'})
```

```python
df
```

```python
#Check Null Values
df.isnull().any()
```

```python
df[df['quote_count'].isna()==1]
```

```python
df['quote_count']=df['quote_count'].fillna(0)
df['reply_count']=df['reply_count'].fillna(0)
df['retweet_count']=df['retweet_count'].fillna(0)
df['favorite_count']=df['favorite_count'].fillna(0)
df['lang']=df['lang'].fillna(0)
df['user_id_str']=df['user_id_str'].fillna(0)
df['conversation_id_str']=df['conversation_id_str'].fillna(0)
df['tweet_url']=df['tweet_url'].fillna(0)
df['username']=df['username'].fillna(0)
```

```python
df[df['tweets']=='Oops']
```

```python
#drop duplicates
df.duplicated()
```

```python
#change DataTypes
df.dtypes
```

```python
#Dropping columns
df=df.drop(columns="conversation_id_str")
```

```python
df
```

```python
#Strip
df["tweets"].str.strip("@")
```

```python
#remove URLs and Special Character
df['tweets']=df['tweets'].str.replace('http[s]://(?:[a-zA-Z]|[0-9][$-_@.&+]|[!*\(\),]|(?:&[0-9A-fA-F][0-9a-fA-F]))+', '')
```

```python
#remove URLs and Special Character
df['tweets']=df['tweets'].str.replace(r"[\"\'\|\?\=\.\<\>\@\#\*\,]", '')
```
