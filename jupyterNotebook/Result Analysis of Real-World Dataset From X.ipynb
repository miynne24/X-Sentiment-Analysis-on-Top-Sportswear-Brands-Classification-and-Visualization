```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'Combined_Brands_Dataset_NB.csv'
data = pd.read_csv(file_path)

# Aggregate the sentiment counts by brand and reorder columns
sentiment_counts = data.groupby(['Brands', 'NaiveBayesSentiment']).size().unstack(fill_value=0).reindex(columns=['Positive', 'Neutral', 'Negative'])

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Define colors for the bars
colors = {'Positive': 'green', 'Neutral': 'yellow', 'Negative': 'red'}

# Plot the grouped bar chart
sentiment_counts.plot(kind='bar', color=[colors.get(x) for x in sentiment_counts.columns], ax=ax, edgecolor='black')

# Add totals on top of each bar
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

# Set the labels and title
ax.set_xlabel('Brands')
ax.set_ylabel('Total Sentiment')
ax.set_title('Sentiment Analysis of Sportswear Brands')
ax.legend(title='Sentiment')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

# Improve layout
plt.tight_layout()
sns.despine()

# Display the plot
plt.show()
```
